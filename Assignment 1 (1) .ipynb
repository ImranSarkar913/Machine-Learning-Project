```python
from sklearn.preprocessing import StandardScaler
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x_scaled , y , test_size = 0.30 , random_state = 1)
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rclf.score(xtest , ytest)
```

Ensemble

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(RandomForestClassifier() , x , y , cv = 5)
scores
```

```python
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag_model = BaggingClassifier(
    estimator = RandomForestClassifier(), 
    n_estimators = 100, 
    max_samples = 0.8, 
    oob_score = True,
    random_state = 0
)
bag_model.fit(xtrain, ytrain)
bag_model.oob_score_
```

```python
bag_model.score(xtest , ytest)
```

```python
from sklearn.metrics import roc_curve , auc ,accuracy_score , classification_report , confusion_matrix
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=["No Disease", "Disease"],
            yticklabels=["No Disease", "Disease"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.savefig("plots/CM Random forst.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
print(classification_report(ytest, pred))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
```

```python
acc = accuracy_score(ytest, pred)
prec = precision_score(ytest, pred)
rec = recall_score(ytest, pred)
f1 = f1_score(ytest, pred)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', cbar=False,
            xticklabels=["No Disease", "Disease"],
            yticklabels=["No Disease", "Disease"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")

plt.subplot(1,2,2)
plt.axis('off')
metrics_text = f"""
Accuracy  : {acc:.2f}
Precision : {prec:.2f}
Recall    : {rec:.2f}
F1-score  : {f1:.2f}
"""
plt.text(0.1, 0.5, metrics_text, fontsize=14, fontweight='bold', verticalalignment='center')

plt.suptitle("Model Evaluation Summary Random Forest", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig("plots/RF CR Random forst.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
fpr, tpr, thresholds = roc_curve(ytest, pred)
roc_auc = auc(fpr, tpr)
```

ROC Curve

```python
plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='green', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Random Forest ROC Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.savefig("plots/RF Roc.png", dpi=300, bbox_inches='tight')
plt.show()
```
